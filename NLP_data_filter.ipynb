```python
from datasets import load_dataset
import pandas as pd
```

```python
dataset = load_dataset("amazon_reviews_multi")
```

```python
df={}

df['train'] = dataset['train'].to_pandas()
df['validation'] = dataset['validation'].to_pandas()
df['test'] = dataset['test'].to_pandas()

df['train']
```

```python
for key in df.keys():
  df[key] = df[key][df[key]['language'] == 'en']

  df[key] = df[key][~df[key].product_category.isin(['personal_care_appliances',
                                                    'musical_instruments',
                                                    'luggage',
                                                    'digital_video_download',
                                                    'industrial_supplies',
                                                    'jewelry',
                                                    'baby_product',
                                                    'book',
                                                    'grocery',
                                                    'shoes',
                                                    'digital_ebook_purchase',
                                                    'pet_products',
                                                    'lawn_and_garden',
                                                    'sports',
                                                    'toy',
                                                    'kitchen',
                                                    'drugstore',
                                                    'beauty',
                                                    'other',
                                                    'apparel',
                                                    'automotive'])]
```
